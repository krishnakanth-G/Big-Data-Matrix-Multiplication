```
m = 2 # number of rows in matrix A
n = 2 # number of columns/rows in matrix A/B
p = 2 # number of columns in magtrix B

def flatmapper(indicesAndValue): # function to convert the input line into key value pair 
  li =[]
  indicesAndValue = indicesAndValue.split(",")
  if indicesAndValue[0] == 'A':
    for k in range(0,p):
      li.append(((int(indicesAndValue[1]),k),(indicesAndValue[0],indicesAndValue[2],indicesAndValue[3])))
  if indicesAndValue[0] == 'B':
    for i in range(0,m):
      li.append(((i,int(indicesAndValue[2])),(indicesAndValue[0],indicesAndValue[1],indicesAndValue[3])))
  return li

def mapper(indicesAndValue): # function to Produce output
    lia = []
    lib = []
    sum =0
    for i in indicesAndValue[1]:
      val = ','.join(i).split(',')
      if(val[0] == 'A'):
        lia.insert(int(val[1]), int(val[2]))
      else:
        lib.insert(int(val[1]), int(val[2]))
    for i in range(0,m):
      for j in range(0,p):
        if(i == j):
          sum = sum + lia[i]*lib[j]
    return [indicesAndValue[0],sum]
                
Matrix_in = sc.textFile('/FileStore/tables/Matrix.txt')
Matrix_out = Matrix_in.flatMap(lambda value: flatmapper(value)).groupByKey()\
                        .map(lambda keyvalue:mapper(keyvalue)).sortByKey()
```
